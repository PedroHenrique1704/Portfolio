## Exercício 1: Vestibular

Considere que a os dados gerados na célula abaixo contêm o número de acertos de 100 alunos em um vestibular para um curso de exatas, divididas pelos respectivos assuntos. Considere que cada assunto possui um número de questões conforme a tabela abaixo:

| assunto | número de questões |
|:---:|:---:|
| Matemática | 24 |
| Português | 18 |
| Geografia | 8 |
| Inglês | 8 |
| História | 8 |
| Física | 12 |
| Química | 12 |

Usando os comandos de operações com DataFrames que você aprendeu na Aula 03, calcule:

1. (operações com escalar) Calcule o percentual de acerto dos alunos por assunto.  
2. (operações entre *DataFrames) Calcule o total de acertos de cada aluno.  
3. Calcule o porcentual geral de cada aluno.  
4. Suponha que a nota de corte para a segunda fase seja 45. Quantos alunos tiveram nota maior que 45?  

````python
import pandas as pd
import numpy as np

np.random.seed(42)
df_mat = pd.DataFrame(np.random.randint(24, size=(100, 1)), columns=['Qt_acertos'])

df_por = pd.DataFrame(np.random.randint(18, size=(100, 1)), columns=['Qt_acertos'])

df_geo = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_ing = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_his = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_fis = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])

df_qui = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])
````

````python
# 1) Seu código aqui

df_mat = df_mat.rename(columns={'Qt_acertos': 'Matemática'})
df_por = df_por.rename(columns={'Qt_acertos': 'Português'})
df_geo = df_geo.rename(columns={'Qt_acertos': 'Geografia'})
df_ing = df_ing.rename(columns={'Qt_acertos': 'Inglês'})
df_his = df_his.rename(columns={'Qt_acertos': 'História'})
df_fis = df_fis.rename(columns={'Qt_acertos': 'Física'})
df_qui = df_qui.rename(columns={'Qt_acertos': 'Química'})



df = pd.concat([df_mat, df_por, df_geo, df_ing, df_his, df_fis, df_qui], axis=1)


percentuais_acertos = df.copy()
percentuais_acertos['Matemática'] = percentuais_acertos['Matemática'] / 24 * 100
percentuais_acertos['Português'] = percentuais_acertos['Português'] / 18 * 100
percentuais_acertos['Geografia'] = percentuais_acertos['Geografia'] / 8 * 100
percentuais_acertos['Inglês'] = percentuais_acertos['Inglês'] / 8 * 100
percentuais_acertos['História'] = percentuais_acertos['História'] / 8 * 100
percentuais_acertos['Física'] = percentuais_acertos['Física'] / 12 * 100
percentuais_acertos['Química'] = percentuais_acertos['Química'] / 12 * 100

print(percentuais_acertos)
````

````python
# 2) Seu código aqui

df['Total_Acertos'] = df.drop(columns=['Total_Acertos'], errors='ignore').sum(axis=1)
print(df['Total_Acertos'])
````

````python
# 3) Seu código aqui

total_questoes = 24 + 18 + 8 + 8 + 8 + 12 + 12
df['Percentual_Geral'] = df['Total_Acertos'] / total_questoes * 100
df['Percentual_Geral'] = df['Percentual_Geral'].round(2)

print(df['Percentual_Geral'])
````

````python
# 4.1) Seu código aqui
# Nessa questão eu fiquei na duvida se a nota de corte sendo 45, seria 45 questões ou 45% da prova, pois a prova possui 90 questões.
#45 questões
nota_corte = 45
alunos_acima_corte = df[df['Total_Acertos'] > nota_corte].shape[0]

print(alunos_acima_corte)
````

````python
# 4.2) Seu código aqui
#45% da prova
alunos_acima_corte = df[df['Percentual_Geral'] > nota_corte].shape[0]

print(alunos_acima_corte)
````

Ainda sobre o mesmo banco de dados:

1. Neste vestibular, quem 'zera' em matemática, física ou química está desqualificado. Monte um novo *DataFrame* com os alunos desqualificados por este critério.
2. Quantos são esses alunos?
3. Qual a média desses alunos em história e geografia?
4. Monte um *DataFrame* com os alunos que passaram para a segunda fase. Repare que estes alunos não podem ter sido desqualificados.

````python
# seu código aqui
# Eu to usando os alunos que possuem mais de 45 acertos, pois da exatamente 50% das questões.
# No lugar dos alunos que acertaram apenas 45% das questões o que seria 40.5 questões.

alunos_desqualificados = df[(df['Matemática'] == 0) | (df['Física'] == 0) | (df['Química'] == 0)]

num_alunos_desqualificados = alunos_desqualificados.shape[0]

media_historia = round(alunos_desqualificados['História'].mean(),2)
media_geografia = round(alunos_desqualificados['Geografia'].mean(),2)

alunos_passaram = df[(df['Matemática'] > 0) & (df['Física'] > 0) & (df['Química'] > 0) & (df['Total_Acertos'] > nota_corte)]

print('Alunos dequalificados por zerar matemática, física ou química: ',num_alunos_desqualificados)
print("Média dos alunos desqualificados em História:", media_historia)
print("Média dos alunos desqualificados em Geografia:", media_geografia)
````
