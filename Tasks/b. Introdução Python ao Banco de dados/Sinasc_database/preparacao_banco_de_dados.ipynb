```python
import pandas as pd
import requests
# 1) seu código aqui
sinasc = pd.read_csv('SINASC_RO_2019.csv')
print(sinasc.shape)

numero_registro = sinasc.shape[0]
numero_registro_nao_duplicados = sinasc.drop_duplicates().shape[0]

print('Número de registros: ', numero_registro)
print('Número de registros não duplicados: ', numero_registro_nao_duplicados)

# Não há duplicados
```

```python
# 2) seu código aqui


total = sinasc.isnull().sum()
porcentagem = (sinasc.isnull().sum() / len(sinasc)) * 100

# Mostrando todos os arquivos faltantes
pd.set_option('display.max_rows', None)
print('Quantidade de valores faltantes: \n')
print(total)
print('\nPorcentagem de valores faltantes: \n')
print(porcentagem)

#Voltando para evitar lag ou demora nos testes futuros
pd.reset_option('display.max_rows')
```

```python
# 3.1) Criando o novo dataframe
sinasc_filtrado = sinasc[['LOCNASC', 'IDADEMAE', 'ESTCIVMAE', 'ESCMAE', 'QTDFILVIVO', 'GESTACAO', 'GRAVIDEZ', 'CONSULTAS', 'APGAR5']]

numero_registro = sinasc_filtrado.shape[0]
numero_registro_nao_duplicados = sinasc_filtrado.drop_duplicates().shape[0]

print('Número de registros: ', numero_registro)
print('Número de registros não duplicados: ', numero_registro_nao_duplicados)

sinasc_filtrado
```

```python
# 3.2) filtrando os dados faltantes no novo dataframe
total = sinasc_filtrado.isnull().sum()
porcentagem = (sinasc_filtrado.isnull().sum() / len(sinasc_filtrado)) * 100
print('Quantidade de valores faltantes: \n')
print(total)
print('\nPorcentagem de valores faltantes: \n')
print(porcentagem)
```

```python
# 4) seu código aqui
sinasc_filtrado = sinasc_filtrado.dropna(subset=['APGAR5'])

total = sinasc_filtrado.isnull().sum()
porcentagem = (sinasc_filtrado.isnull().sum() / len(sinasc_filtrado)) * 100
print('Quantidade de valores faltantes depois da exclusão: \n')
print(total)
print('\nPorcentagem de valores faltantes depois da exclusão: \n')
print(porcentagem)
```

```python
# 5) seu código aqui
sinasc_filtrado.loc[:, ['ESTCIVMAE', 'CONSULTAS']] = sinasc_filtrado[['ESTCIVMAE', 'CONSULTAS']].fillna(9)
```

```python
# 6) Seu código aqui
sinasc_filtrado.loc[:,'QTDFILVIVO'] = sinasc_filtrado['QTDFILVIVO'].fillna(0)
```

```python
# 7.1) Verificação de colunas

#
contagem_coluna = sinasc_filtrado['ESCMAE'].value_counts()
print('\n',contagem_coluna)

contagem_coluna = sinasc_filtrado['GESTACAO'].value_counts()
print('\n',contagem_coluna)

contagem_coluna = sinasc_filtrado['GRAVIDEZ'].value_counts()
print('\n',contagem_coluna)

#Todas as colunas que possuem dados faltantes possuem nela um dado que se repete em mais de 50% do dataframe
#A fim de não atrapalhar o tratamento de dados os dados faltantes serão atribuidos como os dados que mais se repetem em cada coluna
```

```python
# 7.2) inserindo os valores
colunas_com_faltantes = sinasc_filtrado.columns[sinasc_filtrado.isnull().any()]

for coluna in colunas_com_faltantes:
    valor_mais_comum = sinasc_filtrado[coluna].mode()[0]
    sinasc_filtrado.loc[sinasc_filtrado[coluna].isnull(), coluna] = valor_mais_comum

# Conferindo

total = sinasc_filtrado.isnull().sum()
porcentagem = (sinasc_filtrado.isnull().sum() / len(sinasc_filtrado)) * 100
print('Quantidade de valores faltantes depois da modificação: \n')
print(total)
print('\nPorcentagem de valores faltantes depois da modificação: \n')
print(porcentagem)
```

```python
# 7.3 ) Conferindo as colunas após os valores NaN serem modificados

contagem_coluna = sinasc_filtrado['ESCMAE'].value_counts()
print('\n',contagem_coluna)

contagem_coluna = sinasc_filtrado['GESTACAO'].value_counts()
print('\n',contagem_coluna)

contagem_coluna = sinasc_filtrado['GRAVIDEZ'].value_counts()
print('\n',contagem_coluna)
```

8. O Apgar possui uma classificação indicando se o bebê passou por asfixia:
- Entre 8 e 10 está em uma faixa 'normal'.
- Entre 6 e 7, significa que o recém-nascido passou por 'asfixia leve'.
- Entre 4 e 5 significa 'asfixia moderada'.
- Entre 0 e 3 significa 'asfixia severa'.



```python
# 8) seu código aqui

sinasc_filtrado.loc[(sinasc_filtrado['APGAR5'] > 7) & (sinasc_filtrado['APGAR5'] < 11), 'APGAR5_cat'] = 'normal'
sinasc_filtrado.loc[(sinasc_filtrado['APGAR5'] > 5) & (sinasc_filtrado['APGAR5'] < 8), 'APGAR5_cat'] = 'asfixia leve'
sinasc_filtrado.loc[(sinasc_filtrado['APGAR5'] > 3) & (sinasc_filtrado['APGAR5'] < 6), 'APGAR5_cat'] = 'asfixia moderada'
sinasc_filtrado.loc[(sinasc_filtrado['APGAR5'] >= 0) & (sinasc_filtrado['APGAR5'] < 4), 'APGAR5_cat'] = 'asfixia severa'

contagem_coluna = sinasc_filtrado['APGAR5_cat'].value_counts()
print(contagem_coluna, '\n')

print(sinasc_filtrado)
```

```python
## 9) seu código aqui
sinasc_filtrado = sinasc_filtrado.rename(columns=lambda x: x.lower().replace(' ', '_'))

print(sinasc_filtrado.columns)
```
