````python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('previsao_de_renda.csv')
df.head()
````

````python
sns.pointplot(x='posse_de_imovel', y='renda', data=df,errorbar=('ci',95))
````

````python
sns.pointplot(x='posse_de_veiculo', y='renda', hue='sexo', data=df, errorbar=('ci',95))
````

**Insights**

1. Como visto no primeiro gráfico, em média pessoas com maior renda tem mais possibilidades de possuir um carro.

2. Porem quando separamos por sexo, vimos que homens com renda menores tem mais chances de possuirem carros e mulheres com renda maiores tem mais chance de possuirem carros.
