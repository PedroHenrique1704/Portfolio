```python
# 1)

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv('sinasc_RO_2019.csv')
df.columns
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
# 2.1) Não compreendi a questão pois o gráfico ficou meio estranho, fiz um painel logo abaixo.

df_melt = pd.melt(df, value_vars=['APGAR1', 'APGAR5', 'SEXO', 'GESTACAO', 'ESTCIVMAE'], var_name='Variable', value_name='Value')

df_melt= df_melt.sort_values(by=['Variable', 'Value'])

plt.figure(figsize=(15, 10))
sns.countplot(x='Value', hue='Variable', data=df_melt)
plt.xlabel('Valores')
plt.ylabel('Frequencia')
plt.title('Contagem de diferentes variaveis')
plt.legend(title='Variaveis')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
# 2.2) Painel

fig, ax = plt.subplots(3, 2, figsize=(10, 10))

set_ylabels = ['Contagem']

sns.countplot(x='APGAR1', data=df, ax=ax[0][0])
sns.countplot(x='APGAR5', data=df, ax=ax[0][1])
sns.countplot(x='SEXO', data=df, ax=ax[1][0])
sns.countplot(x='GESTACAO', data=df, ax=ax[1][1])
sns.countplot(x='ESTCIVMAE', data=df, ax=ax[2][0])


plt.title('Contagem de diferentes variaveis')

ax[0][0].set_ylabel('Frequencia')
ax[0][1].set_ylabel('Frequencia')
ax[1][0].set_ylabel('Frequencia')
ax[1][1].set_ylabel('Frequencia')
ax[2][0].set_ylabel('Frequencia')


fig.delaxes(ax[2][1])

plt.subplots_adjust(hspace=0.4, wspace=0.4)

ax[1][1].tick_params(axis='x', rotation=30)



plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
# 3)
sns.barplot(x='CONSULTAS', y='APGAR5', data=df)
plt.title('APGAR5 médio por CONSULTAS')
plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
# 4)

fig, ax = plt.subplots(2, 2, figsize=(10, 10))

set_ylabels = ['Contagem']

sns.countplot(x='CONSULTAS', data=df, ax=ax[1][1])
sns.countplot(x='SEXO', data=df, ax=ax[0][0])
sns.countplot(x='GESTACAO', data=df, ax=ax[0][1])
sns.countplot(x='ESTCIVMAE', data=df, ax=ax[1][0])


ax[0][0].set_ylabel('Frequencia')
ax[0][1].set_ylabel('Frequencia')
ax[1][0].set_ylabel('Frequencia')
ax[1][1].set_ylabel('Frequencia')

plt.subplots_adjust(hspace=0.5, wspace=0.3)

ax[0][1].tick_params(axis='x', rotation=30)



plt.show()
```
